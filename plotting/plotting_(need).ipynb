# 1 A simple plot

## 1.1 Let’s look at some code.

### Things to note

```python
from matplotlib import pyplot as plt #matplotlib has alot other stuff, we are just importing pyplot from matplotlib's library of stuff

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6] 
y_1 = [0, 2, 4, 6, 8, 10, 12] #x and y axis need to have the same dimensions
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='blue', linestyle='dashed') #The order of how you specify the keyword arguments (color, linewidth, …) does not matter.
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Hi Derek') #This line adds a title to the plot.
plt.grid(alpha=.25)  #This changes the transparency of the grid.
plt.legend(loc='lower right') #This line adds a legend to the plot at a specfic location
plt.show()
```

## 1.2 Adding another plot

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 3, 15, 9, 12, 24, 18]
# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue') #plot second line
plt.plot(x, y_3, label='Y3 values', color='orange',linestyle="dotted",linewidth=3) #plot third line
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 3, 15, 9, 12, 24, 18]
y_4 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.plot(x, y_3, label='Y3 values', color='orange',linestyle="dotted",linewidth=3) 
plt.errorbar(x, y_4, xerr=x_error, yerr=y_error, # include x and y error bars for the points. 
             label='Y4 with errors', color='green') 
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# 2 Better with NumPy

```python
import numpy as np

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

```python
#One of the advantages of NumPy arrays is that they allow us to generate data-related mathematical functions easily. 
#E.G can plot y = sin x and y = x^2

x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # defining x^2
sin_x = np.sin(x)          # defining sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend() #default upper left?                                
```

## We need another axis!

```python
#Matplotlib offers a variety of ways to have multiple axes.
#The simplest way is to have another y-axis that shares the same x-axis. 
#We can use the command twinx() for this.

x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

```python
x = np.linspace(0, 6, 100) #generates an array of evenly spaced numbers over a specified interval. 
                           #(starting value of the sequence, end value of the sequence, number of samples to generate at fixed interval)
sin_x = np.sin(x)
x2 = x**2

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')
```

# 3 Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150) #We specify resolution by use dots-per-inch (dpi)
plt.savefig('simple-plot.jpeg', dpi=1500) #can save in other modes too! pdf, jpeg
plt.savefig('simple-plot.pdf', dpi=50) #so sharp though as a pdf
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24) #load text from Complete_TAVG_daily.txt, and skip the first 24 rows
date = data[:, 0] #date contains all the values from the first column of the data array
anomaly = data[:, -1] #anomaly contains all the values from the last column of the data array

plt.plot(date, anomaly, alpha=.5) #(x axis, y axis, transparency)
plt.ylim([-4, 4]) #This line sets the y-axis limits of the plot. It restricts the y-axis to the range from -4 to 4.
```

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.25)
plt.ylim([-4, 4])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='red')

#hlines() function needs a y-value and starting (date[0]) and ending values(date[-1]) for x.
```

## 4.2 Styles

```python
plt.style.use('fivethirtyeight') #must set style in the first line
                                 #yucks ugly

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-4, 4])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```

### Task 1

```python
month_convert = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

dummy_file = 'oct-08_singapore.txt'

month = dummy_file[4:6]

new_name = month_convert[dummy_file[0:3]] + "-" + month + "-" +  dummy_file[7:]

print(new_name)
  
```

### Task 2

```python
def rename_my_file(old_file_name):
    
    try:
    
        month = old_file_name[4:6]

        new_name = month_convert[old_file_name[0:3]] + "-" + month + "-" + old_file_name[7:]
        
        return(new_name)
    
    except KeyError:
        
        print("file has already been converted / invalid input")
        
        
   
```

### Task 4

```python
import os
import glob
import shutil

txt_files = glob.glob('*/*.txt')

for file in txt_files:
        original_path = os.path.join(file)
        new_name_yay2 = rename_my_file(file.split(os.path.sep)[-1]) 
                        #.split(os.path.sep) splits the file path into a list of components using the separator "/"
                        #[-1] retrieves the last element of the list obtained from the split operation, which corresponds to the filename itself.
       
        if new_name_yay2==None:

            continue
            
        new_path = os.path.join('os-collaboration-exercise-data', new_name_yay2)
        shutil.move(original_path, new_path)
        
        print(f'renamed from {original_path} to {new_path}')
        
```

### Task 5

```python
country_lst = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for country in country_lst:
    
    new_country_path = os.path.join('os-collaboration-exercise-data', country)
    
    try:
        os.mkdir(new_country_path)
        print(f'{new_country_path} has been created')
        
    except FileExistsError:
        print(f'{new_country_path} already exists; skipping creation.')
```

### Task 6

```python
cairo = glob.glob('os-collaboration-exercise-data/*cairo.txt')

for i in cairo:
    current_path_of_cairo = os.path.join('os-collaboration-exercise-data',i.split(os.path.sep)[-1])
    new_path_of_cairo = os.path.join('os-collaboration-exercise-data',"Cairo", i.split(os.path.sep)[-1])
    
    if os.path.exists(new_path_of_cairo):
        print(f'{new_path_of_cairo} already exists; skipping creation.')
        
    else:
        shutil.move(current_path_of_cairo, new_path_of_cairo)
        print(f'moved from {current_path_of_cairo} to {new_path_of_cairo}')
```

### Task 7

```python

for j in country_lst:
    
    country_file = f"os-collaboration-exercise-data/*{j.lower()}.txt"
    country = glob.glob(country_file)
    
    for i in country:
        
        current_path_of_country = os.path.join('os-collaboration-exercise-data',i.split(os.path.sep)[-1])
        new_path_of_country = os.path.join('os-collaboration-exercise-data',j,i.split(os.path.sep)[-1])
        
        if os.path.exists(new_path_of_country):
            
            print(f'{new_path_of_country} already exists; skipping creation.')
        
        else:
            shutil.move(current_path_of_country, new_path_of_country)
            print(f'moved from {current_path_of_country} to {new_path_of_country}')
        
```
